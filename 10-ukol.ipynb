```python
import pandas
```

Načtení dat o zaměstnancích z CSV:

```python
zam_praha = pandas.read_csv('zam_praha.csv')
zam_plzen = pandas.read_csv('zam_plzen.csv')
zam_liberec = pandas.read_csv('zam_liberec.csv')
```

Přidání nového sloupce 'mesto' do každé tabulky:

```python
zam_praha['mesto'] = 'Praha'
zam_plzen['mesto'] = 'Plzen'
zam_liberec['mesto'] = 'Liberec'
```

Nová tabulka se všemi zaměstnanci:

```python
zamestnanci = pandas.concat([zam_praha, zam_plzen, zam_liberec])
```

Načtení platů z CSV souboru a propojení s tabulkou zaměstnanců:

```python
platy = pandas.read_csv('platy_2021_02.csv')
platy.head()
```

```python
platy_zamestnancu = pandas.merge(zamestnanci, platy, on=['cislo_zamestnance'], how="outer")
platy_zamestnancu
```

Porovnání rozeměrů tabulek před a po spojení:


```python
zamestnanci.shape
```

```python
platy_zamestnancu.shape
```

Chybějící plat u zaměstnance za únor znamená, že již není aktivním zam.:

```python
platy_zamestnancu[platy_zamestnancu['plat'].isna()]
```

Aktivní zaměstnanci firmy:

```python
platy_zamestnancu.dropna(subset=['plat'])
```

Průměrný plat zaměstnanců dle jednotlivých měst:

```python
prumerny_plat = platy_zamestnancu.dropna(subset=['plat'])
```

```python
prumerny_plat.groupby("mesto")["plat"].mean()
```
